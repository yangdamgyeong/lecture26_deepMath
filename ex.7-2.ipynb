```python
import numpy as np

def step_function(x):
    if x > 0:
        return 1
    else:
        return 0

def run_perceptron(inputs, weights, bias):
    z = np.dot(inputs, weights) + bias
    return step_function(z)
```

```python
n = 3
inputs = np.random.rand(n)
weights = np.random.rand(n)
bias = np.random.rand(1)
```

```python
result = run_perceptron(inputs, weights, bias)

print("입력:", inputs)
print("결과:", result)
```

```python
import numpy as np

# 1. 활성화 함수: Step Function (계단 함수)
# 합계가 0보다 크면 1, 아니면 0을 내보냅니다.
def step_function(x):
    return 1 if x > 0 else 0

# 2. 설정값 준비
n = 5  # 입력(x)의 개수를 원하는 만큼 설정하세요.

# x 입력값: 0~1 사이에서 n개 랜덤 생성
x = np.random.rand(n)

# w 가중치: 초기값은 0~1 사이에서 n개 랜덤 생성
w = np.random.rand(n)

# b 편향(bias): 0~1 사이에서 1개 랜덤 생성
b = np.random.rand(1)

# 3. 퍼셉트론 연산
# (입력값 * 가중치)를 모두 더한 뒤 편향(b)을 더합니다.
# np.dot을 쓰면 x1*w1 + x2*w2 + ... + xn*wn 계산이 한 번에 끝납니다.
sum_value = np.dot(x, w) + b

# 4. 활성화 함수 적용 (최종 출력 y)
y = step_function(sum_value)

# --- 결과 확인 ---
print(f"입력값(x): {x}")
print(f"가중치(w): {w}")
print(f"편향(b): {b}")
print("-" * 30)
print(f"가중합(sum): {sum_value}")
print(f"최종 출력(y): {y}")
```

```python
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0

def run_perceptron(n):
    # 입력값(x), 가중치(w)를 n개만큼 0~1 사이 랜덤 생성
    # (랜덤 생성할 때만 넘파이를 쓰고, 계산할 때는 리스트로 변환해서 씁니다)
    x = list(np.random.rand(n))
    w = list(np.random.rand(n))
    b = np.random.rand(1)[0] 
```

```python
total_sum = 0
    
# 0부터 n-1까지 반복하면서 x와 w를 하나씩 곱해서 누적
for i in range(n):
     total_sum = total_sum + (x[i] * w[i])
        
# 마지막에 편향(b)을 더해줍니다.
total_sum = total_sum + b
    
y = step_function(total_sum)

run_perceptron(3)
```

```python
print(f"입력값(x): {x}")
print(f"가중치(w): {w}")
print(f"편향(b): {b}")
print(f"가중합 결과(total_sum): {total_sum}")
print(f"최종 출력(y): {y}")
```
