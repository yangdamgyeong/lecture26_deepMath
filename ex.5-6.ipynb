```python
import pandas as pd
from numpy.linalg import inv
import numpy as np
```

```python
df = pd.read_csv('single.csv', delimiter=",")
```

```python
df
```

```python
X = df.values[:,:-1].flatten()
```

```python
X_1 = np.vstack([X, np.ones(len(X))]).T
```

```python
Y = df.values[:, -1]
```

```python
b = inv(X_1.transpose() @ X_1) @ (X_1.transpose() @ Y)
print(b)
```

```python
y_predict = X_1.dot(b)
```
