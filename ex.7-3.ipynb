```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
all_data = pd.read_csv("simple2.csv")
```

```python
all_data.iloc[:,0:3].max()
```

```python
all_inputs = (all_data.iloc[:, 0:3].values / 255.0)
all_outputs = all_data.iloc[:, -1].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(all_inputs, all_outputs, test_size = 1/3)
n = X_train.shape[0]
```

```python
X_train
```

```python
w_hidden = np.random.rand(3,3)
w_output = np.random.rand(1,3)
```

```python
w_hidden
```

```python
w_output
```

```python
b_hidden = np.random.rand(3,1)
b_output = np.random.rand(1,1)
```

```python
relu = lambda x: np.maximum(x, 0)
logistic = lambda x: 1 / (1 + np.exp(-x))
```

```python
def forward_prop(X):
    Z1 = w_hidden @ X + b_hidden
    A1 = relu(Z1)
    Z2 = w_output @ A1 + b_output
    A2 = logistic(Z2)
    return Z1, A1, Z2, A2
```

```python
train_predictions = forward_prop(X_train.transpose())[3]
```

```python
train_comparisons = np.equal((train_predictions >= 0.5).flatten().astype(int), Y_train)
```

```python
accuracy = sum(train_comparisons.astype(int) / X_train.shape[0])
print("정확도: ", accuracy)
```
